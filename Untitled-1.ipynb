```python
import ipywidgets as widgets
from IPython.display import display
import fitz  # PyMuPDF for PDF processing
import pandas as pd
import re

# Widget to upload PDF file and a button to process
upload_widget = widgets.FileUpload(accept='.pdf', multiple=False)
process_button = widgets.Button(description="Process PDF")

# Display widgets
display(upload_widget, process_button)

# Function to process the uploaded file
def process_pdf(uploaded_file):
    try:
        # Extract the file bytes directly from the uploaded widget
        for file_info in uploaded_file:
            # Correctly access the content from the Bunch object
            pdf_data = uploaded_file[file_info]['content']
        
        # Load the PDF from bytes
        pdf_file = fitz.open(stream=pdf_data, filetype="pdf")
        
        # List to hold the extracted module titles
        module_titles = []
        
        # Regular expression pattern to identify the module titles
        title_pattern = re.compile(r'^\d+\.\s+(.+)$')
        
        # Loop through each page and extract text
        for page_num in range(pdf_file.page_count):
            page = pdf_file[page_num]
            text = page.get_text("text")
            
            # Split text by lines and search for module titles
            for line in text.split('\n'):
                match = title_pattern.match(line)
                if match:
                    # Extract and clean the title of the module
                    module_titles.append(match.group(1).strip())
                    
        # Create a DataFrame to store the titles
        titles_df = pd.DataFrame(module_titles, columns=['Title of the Module'])
        display(titles_df)
    except Exception as e:
        print(f"An error occurred: {e}")

# Function to handle button click
def on_button_click(b):
    if upload_widget.value:
        # Pass the uploaded file value correctly to the processing function
        process_pdf(upload_widget.value)
    else:
        print("Please upload a PDF file before processing.")

# Link button click to the function
process_button.on_click(on_button_click)
```
